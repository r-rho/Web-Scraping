**Single block of code for extracting the quotes from KDNuggets**

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth',100,'display.max_rows',200)
import time
import requests
import random
#reading urls
urls = [f'https://www.kdnuggets.com/2017/05/42-essential-quotes-data-science-thought-leaders.html/{i}' for i in range(1,3)]
urls

#avoiding webscraping detection 
rate = [i/10 for i in range(10)]
time.sleep(random.choice(rate))

#Variables to store the initial data from website
a=[]
q=[]
for url in urls:
    page = requests.get(url)
    soup=bs(page.content,'html.parser')
    a.extend([i.text for i in soup.find_all('a')]) #fetch the Author names from website, under the tag <a>
    q.extend([i.text for i in soup.find_all('ol')]) #fetch the Quotes from website, under the tag <ol> 
    time.sleep(random.choice(rate))
#Remove \n (new line) and storing it to a temporary variable 
length=len(q)
qn=[]
for i in range(length):
    qn.append(q[i].strip())
    
#Slicing the required data from the list
quote = qn[0:22]+qn[26:46]

#Remove three unwanted strings within the author data
g='here'
while g in a: a.remove(g)
a.remove(' here')
a.remove('here.')

#slicing the required author names to a variable
authors=a[43:65]+a[183:204]
authors[40:42] = [' and '.join(authors[40:42])]

quotes = pd.DataFrame({"Quotes":quote,'Author':authors})

quotes
```
